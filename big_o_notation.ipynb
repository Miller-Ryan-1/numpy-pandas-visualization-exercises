```python
#5
import matplotlib.pyplot as plt
import math
```

```python
n = range(1,10)
y1 = [1 for n in n]
y2 = [math.log(n) for n in n]
y3 = [n for n in n]
y4 = [n*math.log(n) for n in n]
y5 = [n**2 for n in n]
y6 = [2**n for n in n]
y7 = [math.factorial(n) for n in n]
y8 = [n**n for n in n]

plt.figure(figsize= (8,12))

plt.title('Big O Notation')

plt.xlabel('Elements')
plt.ylabel('Operations')

plt.plot(n,y1)
plt.plot(n,y2)
plt.plot(n,y3)
plt.plot(n,y4)
plt.plot(n,y5)
plt.plot(n,y6)
plt.plot(n,y7)
plt.plot(n,y8)

plt.xlim(1,5)
plt.ylim(0,50)

plt.legend(['$O(1)$','$O(log(n))$','$O(n)$','$O(n*log(n))$','$O(n^2)$','$O(2^n)$','$O(n!)$','$O(n^n)$'])

plt.show()
```
