```python
import matplotlib.pyplot as plt
```

```python
#1. 
x = range(-10,11)
y = [(x**2 - x + 2) for x in x]

plt.plot(x,y)
plt.text(-1.6,-.6,('0,0'))
```

```python
#2
x = range(0,11)
y1 = [x**(1/2) for x in x]
y2 = [x**3 for x in x]
y3 = [2**x for x in x]
y4 = [1/(x+1) for x in x]

plt.scatter(x,y1)
```

```python
#3
plt.figure(figsize = (10,10))

plt.subplot(2,2,1)
plt.scatter(x,y1)

plt.subplot(2,2,2)
plt.scatter(x,y2)

plt.subplot(2,2,3)
plt.scatter(x,y3)

plt.subplot(2,2,4)
plt.scatter(x,y4)
```

```python
#4
plt.figure(figsize = (10,8))

plt.subplot(2,2,1)
plt.scatter(x,y1, color = 'red')

plt.subplot(2,2,2)
plt.scatter(x,y2, color = 'blue')

plt.subplot(2,2,3)
plt.scatter(x,y3, color = 'green')

plt.subplot(2,2,4)
plt.scatter(x,y4, color = '#4DDDD2')

plt.suptitle('Subplots of Various Equations', fontsize = 20)
```
